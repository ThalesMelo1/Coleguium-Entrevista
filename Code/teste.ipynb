```python
import matplotlib.pyplot as plt
import numpy as np

# Exemplo de dados (substitua pelos seus)
unidades = ['Unidade A', 'Unidade B', 'Unidade C', 'Unidade D', 'Unidade E']
rematricula = [55, 53, 50, 60, 45]
nova_matricula = [5, 8, 10, 7, 6]

x = np.arange(len(unidades))
bar_width = 0.4
overlap = 0.15  # Quanto as barras vão se sobrepor

fig, ax = plt.subplots(figsize=(10, 6))

# Barra 1 - Rematrícula
ax.bar(x, rematricula, width=bar_width, color='royalblue', label='Rematrícula')

# Barra 2 - Nova Matrícula (ligeiramente sobreposta)
ax.bar(x + overlap, nova_matricula, width=bar_width, color='tomato', label='Nova Matrícula', alpha=0.9)

# Ajustes de eixos e rótulos
ax.set_xticks(x + overlap / 2)
ax.set_xticklabels(unidades, rotation=45, ha='right')
ax.set_ylabel('Quantidade')
ax.set_title('Rematrículas x Novas Matrículas por Unidade')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
